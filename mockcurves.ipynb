```python
from pytransit import create_mock_light_curve
```

USE THIS PART IF YOU WANT TO CREATE A SINGLE CURVE:

```python
import os
os.makedirs("data",exist_ok=True)
T=42       #EXPOSITION TIME
period=2.3 #TRANSIT PERIOD
time, flux,params= create_mock_light_curve(texp=T, passband='TESS', noise=3e-4, transit_pars={'period':period})
file = open("data/mockcurve_P"+str(time.size)+"_p"+str(params['period'])+".tess","w")
file.write(str(time.size))
file.write(str(time))
file.write(str(flux))
file.close()
```

```python
import seaborn as sb
from matplotlib.pyplot import subplots, setp

with sb.axes_style('white'):
    fig, ax = subplots()
    ax.plot(time, flux, '.')
    setp(ax, xlim=time[[0,-1]], title='Lightcurve with '+str(time.size)+' points',
         xlabel='Time [d]', ylabel='Normalised flux')
    fig.tight_layout()
```

USE THIS PART IF YOU WANT TO CREATE SEVERAL CURVES:

```python
import os
os.makedirs("data",exist_ok=True))
import numpy as np
for period in np.linspace(1.0, 400.0, num=400):
  for T in [0.04,0.08,0.16,0.32,0.64,1.28, 2.56, 5.12, 10.24, 20.48, 41.96]:
    time, flux,params= create_mock_light_curve(texp=T, passband='TESS', noise=3e-4, transit_pars={'period':period})
    file = open("data/mockcurve_P"+str(time.size)+"_p"+str(params['period'])+".tess","a")
    file = open("data/mockcurve_P"+str(time.size)+"_p"+str(params['period'])+".tess","w")
    file.write(str(time.size))
    file.write(str(time))
    file.write(str(flux))
    file.close()
```
